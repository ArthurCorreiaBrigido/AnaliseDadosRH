- ### 3. *Análise Exploratória Inicial (EDA)*

```python
# --- Distribuição das idades ---
plt.figure(figsize=(10,6))
plt.hist(df['Idade'], bins=20, edgecolor='black', alpha=0.85)
plt.xlabel('Idade', fontsize=12)
plt.ylabel('Frequência', fontsize=12)
plt.title('Distribuição das Idades dos Funcionários', fontsize=14, fontweight='bold')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# --- Frequência das variáveis categóricas ---
categorias = df.select_dtypes(include=['object', 'category']).columns.tolist()
print("Variáveis categóricas encontradas:", categorias)

for col in categorias:
    plt.figure(figsize=(10,6))
    ax = sns.countplot(data=df, x=col, edgecolor='black', pallette='Paired')

    plt.title(f'Frequência de {col}', fontsize=14, fontweight='bold')
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Frequência', fontsize=12)
    plt.xticks(rotation=45)

    for p in ax.patches:
        value = int(p.get_height())
        ax.annotate(f'{value}', (p.get_x() + p.get_width() / 2, p.get_height()),
                    ha='center', va='bottom', fontsize=12,
                    xytext=(0, 5), textcoords='offset points')
        
    plt.grid(axis='y', alpha=0.25)
    plt.tight_layout()
    plt.show()
```

```python
# ---- Distribuição do Nível de Satisfação ----
plt.figure(figsize=(10,6))
ax = sns.countplot(data=df, x='Nivel_Satisfacao_Trabalho', edgecolor='black', palette='Set2')

plt.title('Distribuição da Satisfação no Trabalho', fontsize=14, fontweight='bold')
plt.xlabel('Nível de Satisfação', fontsize=12)
plt.ylabel('Quantidade de Funcionários', fontsize=12)

# Rótulos
for p in ax.patches:
    value = int(p.get_height())
    ax.annotate(f'{value}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=11,
                xytext=(0, 5), textcoords='offset points')

plt.grid(axis='y', alpha=0.25)
plt.tight_layout()
plt.show()
```

    ### 1. *Qual o total de funcionários atualmente na empresa?*

### 2. *Qual o tempo médio de experiência dos funcionários (em anos)?* 

```python
tempo_medio_experiencia = df['Anos_Experiencia'].mean()
tempo_medio_experiencia
```

### 3. *Qual o total e percentual de funcionários do gênero masculino e feminino?*

```python
genero_total = df['Genero'].value_counts()
genero_percentual = df['Genero'].value_counts(normalize=True) * 100

pd.DataFrame({'Total': genero_total, 'Percentual (%)': genero_percentual})
```

### 4. *Qual a média salarial mensal?* 


```python
media_salario = df['Salario_Mensal'].mean()
media_salario
```

### 5. *Qual o total de funcionários por função?*

```python
funcao_total = df['Funcao'].value_counts()
funcao_total
```

### 6. *Qual o percentual de funcionários disponíveis para fazer hora extra?*

```python
percentual_hora_extra = df['Disponivel_Hora_Extra'].value_counts(normalize=True) * 100
percentual_hora_extra
```

### 7. *Qual o nível de envolvimento dos funcionários no trabalho considerando 4 categorias: Ruim, Baixo, Médio e Alto?*

```python
# Conversão da escala numérica para categorias
mapa = {1: 'Ruim', 2: 'Baixo', 3: 'Médio', 4: 'Alto'}
df['Categoria_Envolvimento'] = df['Indice_Envolvimento_Trabalho'].map(mapa)

envolvimento_total = df['Categoria_Envolvimento'].value_counts()
envolvimento_percentual = df['Categoria_Envolvimento'].value_counts(normalize=True) * 100

pd.DataFrame({'Total': envolvimento_total, 'Percentual (%)': envolvimento_percentual})
```

### 8. *Qual o total e o percentual de funcionários que devem receber promoção? Considere a coluna “Anos Desde a Última Promoção" com a seguinte regra: Se o funcionário tiver 5 anos ou mais desde a última promoção, deve ter a promoção considerada. Caso contrário, a promoção não deve ser considerada agora.*

```python
df['Elegivel_Promocao'] = df['Anos_Desde_Ultima_Promocao'] >= 5

total_promo = df['Elegivel_Promocao'].sum()
percentual_promo = (df['Elegivel_Promocao'].mean()) * 100

pd.DataFrame({'Total elegível': [total_promo], 'Percentual (%)': [percentual_promo]})
```
